## Data Cleaning

```python
# 1. Handling Missing Values
dataframes = {
    "Customers": customers_df,
    "Products": products_df,
    "Stores": stores_df,
    "Sales": sales_df,
    "Inventory": inventory_df
}
```

```python
for name, df in dataframes.items():
    print(f"--- Missing Values in {name} ---")
    print(df.isnull().sum())
    print("\n")
```

```python
#For numerical columns with missing values, replace them with the column mean
mean_weight = products_df['weight'].mean()
products_df['weight'] = products_df['weight'].fillna(mean_weight)
```

```python
print(products_df['weight'].isnull().sum())
```

```python
mean_reorder = inventory_df['reorder_level'].mean()
inventory_df['reorder_level'] = inventory_df['reorder_level'].fillna(mean_weight)
```

```python
print(inventory_df['reorder_level'].isnull().sum())
```

```python
#For categorical columns with missing values, replace them with the most frequent value
most_frequent_value = sales_df['payment_method'].mode()[0]
sales_df['payment_method'] = sales_df['payment_method'].fillna(most_frequent_value)
```

```python
print(sales_df['payment_method'].isnull().sum())
```

```python
most_frequent_value = customers_df['zip_code'].mode()[0]
customers_df['zip_code'] = customers_df['zip_code'].fillna(most_frequent_value)
```

```python
print(customers_df['zip_code'].isnull().sum())
```

```python
#3 For date columns with missing values, use forward fill or backward fill as appropriate
sales_df['date'] = sales_df['date'].ffill()
```

```python
print(sales_df['date'].isnull().sum())
```

```python
inventory_df['last_restock_date'] = inventory_df['last_restock_date'].ffill()
```

```python
print(inventory_df['last_restock_date'].isnull().sum())
```

```python
customers_df['registration_date'] = customers_df['registration_date'].ffill()
```

```python
print(customers_df['registration_date'].isnull().sum())
```
