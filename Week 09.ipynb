## Part 2: GroupBy and Aggregation

```python
#Group products by category and calculate the average price for each category
category_averages = products_df.groupby('category')['price'].mean()
```

```python
print(category_averages)
```

```python
#Count how many products are in each category
category_counts = products_df['category'].value_counts()
print(category_counts)
```

```python
#Group customers by state and find the average age in each state
customers_by_state = customers_df.groupby('state')['age'].mean()
```

```python
print(customers_by_state)
```

```python
#Calculate total spending per customer (sum up all their purchases)
customer_spending = sales_df.groupby('sale_id')['total_amount'].sum()
```

```python
print(customer_spending.sort_values(ascending=False).head())
```

```python
#Find the average transaction amount by day of the week
sales_df['day_of_week'] = sales_df['transaction_date'].dt.day_name()
```

```python
#Determine which product category generates the most total revenue
merged_df = sales_df.merge(products_df, on='product_id')
```

```python
category_revenue = merged_df.groupby('category')['total_amount'].sum()
```

```python
top_category = category_revenue.idxmax()
highest_revenue = category_revenue.max()
print(f"The top category is {top_category} with ${highest_revenue:,.2f} in total revenue.")
```

```python
#Find which state has the most customers
state_count = customers_df['state'].value_counts()
```

```python
print(f" the state with most customers is {top_state} with {customer_total} customers.")
```

```python
#Identify the top 5 customers by total spending
spending_by_id = sales_df.groupby('customer_id')['total_amount'].sum()
```

```python
top_customers = pd.merge(spending_by_id, customers_df, on='customer_id')
```

```python
top_5_spenders = top_customers.sort_values(by= 'total_amount', ascending=False).head(6)
```
